# Histograms and Box Plots

```python
sns.histplot(df['NOx(GT)'], kde=True)   # The kde=True argument adds a kernel density estimate (KDE) line to the plot, showing the estimated probability density function of the data.
df.hist(figsize=(20,20),                #This code creates histograms for all numerical columns in the DataFrame df. Here's a breakdown of the arguments used:
        grid=True,
        bins=20,
        color="maroon",
        legend=True)
plt.show()

```

```python
df.boxplot(figsize=(30,10),
          color="maroon",
          grid=True)
```

# CORRELATION AND PAIR PLOT

```python
# Drop non-numeric columns if they're irrelevant
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# Create the correlation heatmap for numeric columns
plt.figure(figsize=(14, 14))
sns.heatmap(df_numeric.corr(), annot=True, cmap="Greens")
plt.show()

```
